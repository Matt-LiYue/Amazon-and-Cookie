```python
class Solution(object):
    def countAndSay(self, n):
        """
        :type n: int
        :rtype: str
        """
        j = 1
        result = "1"
        while j < n:
            j+=1
            lastchar = None
            newresult = ""
            counter = 1
            for char in result:
                if char != lastchar and lastchar is not None:
                    newresult += str(counter)+lastchar
                    counter = 1
                elif lastchar is not None:
                    counter+=1
                lastchar = char
            newresult += str(counter)+lastchar
            result = newresult
        return str(result)
        
        
        
sol = Solution()
sol.countAndSay(6)
```

```python
import time
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        low, high = 0, len(nums)
        while high-low > 1:
            i = (low+high)/2
            if nums[i] > target:
                high = i
            elif nums[i] < target:
                low = i
            else:
                return i
        if target > nums[low]:
            return low+1
        else:
            return low
        
        
        
        
        
sol = Solution()
a = range(0,5)
print sol.searchInsert(a,7)

```

```python
class Solution(object):
    def findSubstring(self, s, words):
        """
        :type s: str
        :type words: List[str]
        :rtype: List[int]
        """
        sublen = len(words)*len(words[0])
        worddic = {}
        result = []
        for word in words:
            if word in worddic:
                worddic[word] += 1
            else:
                worddic[word] = 1
        for i in range(len(s)-sublen+1):
            tmpdic = worddic.copy()
            j = i
            for j in range(i,i+sublen,len(words[0])):
                wordseg = s[j:j+len(words[0])]
                if wordseg in tmpdic and tmpdic[wordseg]>0:
                    tmpdic[wordseg] -= 1
                else:
                    break
                if j+len(words[0]) == i+sublen:#end
                    result.append(i)
                
        return result
            
        
        
s = "wordgoodgoodgoodbestword"
words = ["word","good","best","good"]
sol = Solution()
sol.findSubstring(s,words)
    
        
```

```python
class Solution(object):
    def nextPermutation(self, nums):
        """
        :type nums: List[int]
        :rtype: void Do not return anything, modify nums in-place instead.
        """
        if len(nums) < 2:
            return
        for i in range(len(nums)-1,0,-1):
            if nums[i] > nums[i-1]:
                for k in range(len(nums)-1,i-1,-1):
                    if nums[k] > nums[i-1]:
                        tmp = nums[k]
                        nums[k] = nums[i-1]
                        nums[i-1] = tmp
                        break
                j = 0
                while i+j < len(nums)-1-j:
                    tmp = nums[i+j]
                    nums[i+j] = nums[len(nums)-1-j]
                    nums[len(nums)-1-j] = tmp
                    j+=1
                return
        nums.sort()
                    
                
        
        
sol = Solution()
nums = [1,2,9,4,2,5]
print nums
for i in range(10):
    sol.nextPermutation(nums)
    print nums
```
